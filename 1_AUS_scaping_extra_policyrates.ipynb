```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
import re
from tqdm import tqdm_notebook as tqdm

#For tokenizing sentences
import nltk
nltk.download('punkt')
```

```python
df = pd.read_csv('df_australia.csv')
df
```

```python
#Additionally extract cash rate
cash_rate_value = r'([\d.]+) per cent' #(to|at)
```

```python
cash_rates = [re.search(cash_rate_value, text).group(1) if re.search(cash_rate_value, text) \
              else None for text in df['Statements']]
```

```python
df['Cash Rate'] = cash_rates
df
```
